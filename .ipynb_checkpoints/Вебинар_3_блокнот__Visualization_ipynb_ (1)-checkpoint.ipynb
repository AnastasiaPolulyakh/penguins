## Описание данных и предобработка
Продолжаем работать с данными по COVID-19 ([источник](https://github.com/datasets/covid-19)). Переменные и их описание:

- `Date` — дата
- `Country/Region` — страна/регион
- `Province/State` — штат/провинция
- `Confirmed` — количество подтверждённых случаев заражения на указанную дату
- `Recovered` — количество выздоровевших на указанную дату
- `Deaths` — количество умерших на указанную дату

## Предобработка
Была выполнена на предыдущем занятии, здесь останется только код.

```python
import numpy as np
import pandas as pd

pd.options.display.float_format = '{:.3f}'.format

data = pd.read_csv('time-series-19-covid-combined.csv')
data.info()
```

Ренейминг и обработка даты:

```python
data_renamed = data.rename(columns=lambda name: name.lower().replace('/', '_'))
data_renamed['date_dt'] = pd.to_datetime(data_renamed['date'], format='%Y-%m-%d')

data_renamed.head()
```

Обработка пропущенных значений и фильтрация территориальных юнитов, не являющихся странами:

```python
not_countries = ['Antarctica', 'Diamond Princess', 'MS Zaandam', 'Summer Olympics 2020', 'Winter Olympics 2022']

data_filtered = data_renamed[
    (data_renamed['country_region'] != 'Canada') & ~(data_renamed['country_region'].isin(not_countries))
].reset_index(drop=True)  # reset_index сбрасывает индексы к дефолтному состоянию, автоматически этого не произойдёт
data_filtered['recovered'] = data_filtered['recovered'].astype(int)

data_filtered.head()
```

Переход от данных в разрезе регионов к данным в разрезе стран:

```python
NUMERIC = ['confirmed', 'recovered', 'deaths']

data_aggregated = data_filtered.groupby(['date', 'date_dt', 'country_region'], as_index=False)[NUMERIC].sum()
data_aggregated.head()
```

Переход от накопленных данных на период к изменению за период:

```python
def get_deltas(country_df: pd.DataFrame) -> pd.DataFrame:
    country_df = country_df.copy()

    for name in NUMERIC:
        country_df[f'{name}_delta'] = country_df[name] - country_df[name].shift(fill_value=0)

    return country_df

data_final = data_aggregated.groupby(['country_region']).apply(get_deltas, include_groups=False).reset_index(level=0).reset_index(drop=True)

for name in NUMERIC:
    data_final.loc[data_final[f'{name}_delta'] < 0, f'{name}_delta'] = 0

data_final.head()
```

Подгрузим другую таблицу — с базовыми индикаторами для некоторых
стран.

```python
def extract_index(index_tuple: tuple) -> str:
    first, second = index_tuple

    if not second:
        return first

    return second
```

```python
indicators = ['Population, total', 'GDP per capita (current US$)', 'Life expectancy at birth, total (years)']

indicators_df = pd.read_excel('P_Popular Indicators.xlsx').drop(columns=['Series Code'])
```

```python
indicators_data = indicators_df[indicators_df['Series Name'].isin(indicators)].reset_index(drop=True)

# melt year columns (WIDE[years] -> LONG[year])
indicators_data_melt = indicators_data.melt(
    id_vars=['Series Name', 'Country Name'],
    value_vars=['2019 [YR2019]', '2020 [YR2020]', '2021 [YR2021]'],
    var_name='year', value_name='value'
)

# pivot Series Name column (LONG[Series Name] -> WIDE[Series Name])
indicators_data_pivot = indicators_data_melt.pivot(
    columns=['Series Name'], index=['Country Name', 'year'], values=['value']
).reset_index()

# drop multiindex
indicators_data_pivot.columns = indicators_data_pivot.columns.map(extract_index)  # we have defined extract_index function earlier

# rename_columns
renamer = {
    'Country Name': 'country_name',
    'GDP per capita (current US$)': 'grp_percap',
    'Life expectancy at birth, total (years)': 'life_exp',
    'Population, total': 'population',
}

indicators_data_pivot.rename(columns=renamer, inplace=True)

# transform variables
indicators_data_pivot['year'] = indicators_data_pivot['year'].str[:4].astype(int)

indicators_data_pivot['grp_percap'] = [value if isinstance(value, float) else np.nan for value in indicators_data_pivot['grp_percap']]
indicators_data_pivot['life_exp'] = [value if isinstance(value, float) else np.nan for value in indicators_data_pivot['life_exp']]
indicators_data_pivot['population'] = [value if isinstance(value, int) else np.nan for value in indicators_data_pivot['population']]

indicators_data_pivot.info()
```

После всех преобразований выполним left join по полям «год» и «имя страны»:

```python
data_final_to_merge = data_final.copy()

data_final_to_merge.insert(0, 'year', data_final['date_dt'].dt.year)
```

```python
left_cols = ['country_region', 'year']
right_cols = ['country_name', 'year']

data_merged = data_final_to_merge.merge(
    indicators_data_pivot,
    left_on=left_cols,
    right_on=right_cols,
    how='left'
).drop(columns=['country_name'])
```

# Теоретическая часть: классификация переменных и алгоритмы

## Классификация шкал переменных
Вспомним, как классифицируются переменные по типу шкалы. Бывают вариации, но в целом различий практически нет:

1. **Номинальная, категориальная шкала.** Переменная описывает категории, которые нельзя упорядочить или сравнить численно
2. **Порядковая, ординальная шкала.** Переменная имеет значения, которые можно упорядочить, но разница между ними не является фиксированной
3. **Интервальная шкала.** Между значениями переменной равные промежутки, но нет истинной нулевой точки. Например, значение температуры по Цельсию
4. **Относительная, пропорциональная шкала.** Шкала имеет истинную нулевую точку, что позволяет сравнивать величины пропорционально

Далее мы будем объединять последние две шкалы в единую категорию — **числовые**.

## Классификация распределения по количеству и типу переменных
**Распределение** может характеризоваться типом и числом переменных, которые в него входят. Мы рассмотрим случаи распределения одной переменной и взаимодействия двух переменных. В зависимости от типа могут быть такие варианты:

1. Одна переменная:
    - числовая
    - категориальная
    - ординальная
2. Две переменных:
    - числовая & числовая
    - числовая & категориальная
    - числовая & ординальная
    - категориальная & ординальная
    - категориальная & категориальная
    - ординальная & ординальная

Бо́льшую часть работы аналитика можно свести к изучению совместного распределения $N$ переменных. Теперь по порядку рассмотрим каждый случай.

Про квантили: https://www.statisticshowto.com/quantile-definition-find-easy-steps/.

Понятие квантиля тесно связано с понятием **функции распределения**:

```python
sample = [0, 4, -2, 5, 9, 8]

# значение функции распределения от x - доля значений выборки, которые меньше или равны данному
# F(-2) = 1 / 6
# F(0) = 2 / 6 = 1 / 3
# F(1) = 2 / 6 = 1 / 3
# F(9) = 6 / 6 = 1
```

```python
# 1 / 6  ->  -2
```

В сущности, это способ "приблизить" [density plot](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.density.html) (его мы рисовать не будем, потому что там надо подбирать функцию сглаживания). В гистограмме многое зависит от выбранного количества бинов (палочек):

Наконец, если вам нужно прикинуть функцию распределения переменной, можно построить эмпирическую функцию распределения (Empirical Cumulative Distribution Function, или ECDF).

**Empirical Cumulative Distribution Function**<br>
Как нетрудно догадаться, строится по аналогии с [классической](https://www.statisticshowto.com/cumulative-distribution-function-cdf/) функцией распределения. CDF:

```python
from scipy import stats

norm_distrib = stats.norm(0, 1)

x = np.linspace(-4, 4, 10000)
f_x = norm_distrib.cdf(x)

quantile0_6 = norm_distrib.ppf(0.6)            # считаем квантиль уровня 0.6
f_quantile0_6 = norm_distrib.cdf(quantile0_6)  # функция распределения от квантиля уровня 0.6 (равна 0.6)

fig, ax = plt.subplots(figsize=(10, 7))

ax.plot(x, f_x)

# рисуем стрелочки
ax.arrow(x=-4, y=f_quantile0_6, dx=(quantile0_6 - (-4)), dy=0, color='red', head_width=0.01, head_length=0.1, length_includes_head=True)
ax.arrow(x=quantile0_6, y=f_quantile0_6, dx=0, dy=-f_quantile0_6, color='red', head_width=0.05, head_length=0.02, length_includes_head=True)

ax.axhline(0, color='black')

ax.set_title('CDF (standard normal distribution)')
ax.set_xlabel('x')
ax.set_ylabel('F(x)');
```

ECDF:

```python
# выборка потенциально из стандартного нормального распределения
some_sample = np.random.normal(0, 1, (1000,))
random_noize = np.random.uniform(-1, 1, (1000,))
some_sample = some_sample * random_noize
```

**Задача.** Напишите функцию, которая принимает на вход массив и возвращает массив, в котором в соответствие каждому элементу изначального массива поставлена доля значений массива, которые меньше или равны данному.

```python
def get_empirical_f_x(array: np.ndarray) -> np.ndarray:  # считаем значение функции распределения не по формуле, а эмпирически
    array_f_x = np.array([(array <= x).mean() for x in array])

    return array_f_x
```

```python
empirical_f_x = get_empirical_f_x(some_sample)
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

ax.scatter(some_sample, empirical_f_x)

ax.set_title('ECDF (suspected standard normal distribution)')
ax.set_xlabel('x')
ax.set_ylabel('Proportion of values X <= x');
```

Ограничения **коэффициента корреляции Пирсона**:

- требование нормальности распределения данных;
- предположение о линейной взаимосвязи;
- влияние выбросов (зависимость от масштаба измерений).

Что можно использовать вместо:

- **коэффициент корреляции Спирмана**:
    - устойчив к выбросам, но требует разных значений;<br>
![](https://cdn3.slideserve.com/6553146/spearman-s-rank-correlation-n.jpg)
[Источник изображения](https://cdn3.slideserve.com/6553146/spearman-s-rank-correlation-n.jpg)

**Перевод**: Ранговая корреляция Спирмена. Коэффициент ранговой корреляции Спирмена - это статистический показатель, показывающий силу взаимосвязи между двумя переменными. Десять - минимальное число, необходимое для того, чтобы ранговый тест Спирмена был валидным
- **тау Кендалла:**
    - минус предложение о линейной взаимосвязи, но все равно она должна быть монотонной;
    - менее чувствителен к силе связи, т. е. почти всегда будет ниже, чем Пирсона или Спирмана;
    - не требует нормальности распределения (вообще не учитывает форму распределения);
- etc.

Выглядит не очень красиво, бины не совпадают. Поправить положение в `matplotlib` можно используя кастомные бины:

```python
def get_bins(series: pd.Series, n=20):
    bins = np.linspace(series.min(), series.max(), n)

    return bins
```

```python
fig, ax = plt.subplots(figsize=(10, 7))

bins = get_bins(indicators_data_pivot2019_2['life_exp'])

for region in world_regions:
    ax.hist(indicators_data_pivot2019_2.loc[indicators_data_pivot2019_2['world_region'] == region, 'life_exp'], edgecolor='black', bins=bins, alpha=.2, label=region)

ax.set_title('life_exp & RainTomorrow histogram')
ax.set_xlabel('life_exp')
ax.set_ylabel('Count')

ax.legend();
```

А вот в `seaborn` все куда проще:)

```python
fig, ax = plt.subplots(figsize=(10, 7))

sns.histplot(x='life_exp', hue='world_region', ax=ax, data=indicators_data_pivot2019_2, alpha=0.4, bins=20)

ax.set_title('life_exp & world_region histogram');
```

Можно построить и т. н. stacked гистограмму (бины кладутся один на другой):

Можно посмотреть как на совместное распределение, так и на долю каждой в каждом бине:

```python
fig, ax = plt.subplots(figsize=(10, 7))

sns.histplot(x='life_exp', hue='world_region', ax=ax, data=indicators_data_pivot2019_2, alpha=0.4, bins=20, multiple='fill')

ax.set_title('life_exp & world_region histogram [filled]');
```

Dodge гистограмма (бины находятся рядом):

```python
fig, ax = plt.subplots(figsize=(10, 7))

sns.histplot(x='life_exp', hue='world_region', ax=ax, data=indicators_data_pivot2019_2, alpha=0.4, bins=20, multiple='dodge')

ax.set_title('life_exp & world_region histogram [dodge]');
```

Гистограммы не всегда просто сравнивать, особенно если категорий много. Совсем другая ситуация с boxplot'ами:

```python
fig, ax = plt.subplots(figsize=(10, 7))

sns.boxplot(x='life_exp', y='world_region', data=indicators_data_pivot2019_2, ax=ax)

ax.set_title('life_exp & world_region boxplot');
```

Boxplot, как можно заметить, отлично подходит для сравнения распределений. Но можно сравнить и сами функции распределения:

```python
fig, ax = plt.subplots(figsize=(10, 7))

for region in world_regions:
    gdp_region = indicators_data_pivot2019_2.loc[indicators_data_pivot2019_2['world_region'] == region, 'life_exp'].values
    f_gdp_region = get_empirical_f_x(gdp_region)

    ax.scatter(gdp_region, f_gdp_region, marker='.', label=region)

ax.set_title('life_exp & world_region ECDF')
ax.set_xlabel('life_exp')
ax.set_ylabel('proportion')

ax.legend();
```

Видно, что функция распределения переменной `life_exp` для региона Африка "набирает плотность" быстрее всех.

Наконец, errorbar позволяет сравнить не распределение, а его характеристики:

```python
life_exp_by_region_means = indicators_data_pivot2019_2.groupby('world_region')['life_exp'].mean().round(2)
life_exp_by_region_stds = indicators_data_pivot2019_2.groupby('world_region')['life_exp'].std().round(2)

fig, ax = plt.subplots(figsize=(10, 7))

ax.bar(life_exp_by_region_means.index, life_exp_by_region_means.values, edgecolor='black')
ax.errorbar(life_exp_by_region_means.index, life_exp_by_region_means.values, yerr=life_exp_by_region_stds, fmt='o', color='red')  # среднее и стандартное отклонение

ax.set_title('life_exp & world_region errobar')
ax.set_xlabel('world_region')
ax.set_ylabel('life_exp');
```

### Числовая & ординальная переменные
##### Метрики
Поскольку ординальная переменная представляет из себя нечто среднее между числовой и категориальной переменной, возможно несколько вариантов. Если такая переменная содержит нечисловые значения, то с ней можно обращаться как с категориальной:

```python
indicators_data_pivot2019_2.groupby(categorical, observed=False)[['life_exp', 'population']].mean().round(2)  # надо контролировать очередность по индексу
```

Если же она содержит целые значения, то можно использовать коэффициенты корреляции:

```python
# переведем население в миллионы
indicators_data_pivot2019_2['population_mln'] = ((indicators_data_pivot2019_2['population'] // 1000000) / 10).round() * 10
indicators_data_pivot2019_2['population_mln'].unique()
```

```python
indicators_data_pivot2019_2['population_mln'].corr(indicators_data_pivot2019_2['grp_percap'], method='spearman')
```

```python
indicators_data_pivot2019_2['population_mln'].corr(indicators_data_pivot2019_2['grp_percap'], method='kendall')
```

#### Способы визуализации
С визуализацией ситуация аналогичная. Если переменная больше "категориальная":

```python
fig, ax = plt.subplots(figsize=(10, 7))

sns.boxplot(x='life_exp', y=categorical, data=indicators_data_pivot2019_2, ax=ax)

ax.set_title('life_exp & GDP categorical boxplot');
```

Если же числовая:

```python
indicators_data_pivot2019_2_filtered = indicators_data_pivot2019_2[indicators_data_pivot2019_2['population_mln'] < 1000].reset_index(drop=True)

spearman_coef = indicators_data_pivot2019_2_filtered['population_mln'].corr(indicators_data_pivot2019_2_filtered['life_exp'], method='spearman')

fig, ax = plt.subplots(figsize=(10, 7))

sns.scatterplot(x='population_mln', y='life_exp', ax=ax, data=indicators_data_pivot2019_2_filtered, alpha=0.4)

ax.set_title(f'Scatterplot of vars population_mln and life_exp\n(Spearman coef is {spearman_coef:.02})')
ax.set_xlabel('population_mln')
ax.set_ylabel('life_exp');
```

Даже если порядковая переменная является числовой, обычно количество уникальных категорий не является большим. В таком случае scatterplot выходит некрасивым (как можно видеть выше). Решение - т. н. `jittering`, или добавление случайного шума по произвольной оси:

```python
indicators_data_pivot2019_2_filtered['population_mln_jittered'] = (
    indicators_data_pivot2019_2_filtered['population_mln'] + np.random.normal(0, 5, indicators_data_pivot2019_2_filtered['population_mln'].size)
)
```

```python
spearman_coef = indicators_data_pivot2019_2_filtered['population_mln'].corr(indicators_data_pivot2019_2_filtered['life_exp'], method='spearman')

fig, ax = plt.subplots(figsize=(10, 7))

sns.scatterplot(x='population_mln_jittered', y='life_exp', ax=ax, data=indicators_data_pivot2019_2_filtered, alpha=0.4)

ax.set_title(f'Scatterplot of vars population_mln_jittered and life_exp\n(Spearman coef is {spearman_coef:.02})')
ax.set_xlabel('population_mln_jittered')
ax.set_ylabel('life_exp');
```

Смотреть на абсолютные значения особенного смысла нет. Нормализуем:

```python
pd.crosstab(categorical, indicators_data_pivot2019_2_filtered['world_region'], normalize=1).round(2)  # по столбцам
```

#### Способы визуализации
Как таковую визуализацию здесь можно не строить, т. к. таблица выше сама по себе является визуализацией. Однако если она громоздкая, можно облегчить ее понимание с помощью heatmap:

```python
crosstab = pd.crosstab(categorical, indicators_data_pivot2019_2_filtered['world_region'])
crosstab_normalized = pd.crosstab(categorical, indicators_data_pivot2019_2_filtered['world_region'], normalize=True).round(2)

fig, ax = plt.subplots(figsize=(6, 5))

sns.heatmap(
    crosstab_normalized,
    ax=ax,
    annot=True,
    vmin=0,
    vmax=1,
    center=1 / crosstab_normalized.size,
    cmap='coolwarm',
    linewidths=1,
    linecolor='white',
    square=True,
);
```

Чем больше значения отличается от теоретически ожидаемого, тем больше цвет отклоняется от среднего в палитре.
