## Про тип `object`
Здесь мы немного прервемся, отдохнем от сложного и поговорим на важную тему. На самом деле `np.ndarray` и `pd.Series` могут хранить значения разных типов - используя тип `object`.

```python
[1.4, '1,4']
```

```python
np.array([1, '1'])
```

```python
weird_array = np.array([1, '1'], dtype='object')
weird_array
```

Такой `array` лишается преимуществ массива (в первую очередь скорости вычислений). Пример:

```python
array_int = np.arange(100000)
array_object = array_int.astype('object')
```

```python
%%timeit

_ = array_int * 2
```

```python
%%timeit

_ = array_object * 2
```

Будьте осторожны с типами данных:)

## Преобразование таблиц: pivot(_table) & melt
Теперь немного поговорим о двух функциях, которые используются при работе с Pandas чаще, чем можно было бы представить.

![](https://i.ytimg.com/vi/AsZPQyjYBhc/maxresdefault.jpg)

Как видно на картинке выше, табличные данные можно хранить в нескольких форматах.

**Long-формат**<br>
В long формате таблицы данные представлены в виде множества строк с меньшим количеством столбцов. Каждая строка обычно содержит одну запись с множеством переменных. В таких таблицах часто один столбец содержит названия переменных, а другой — соответствующие значения.

**Wide-формат**<br>
В wide формате таблицы данные представлены с множеством столбцов и меньшим количеством строк. Каждый столбец представляет отдельную переменную.

Давайте рассмотрим это на практике.

```python
countries = ['US', 'Russia', 'United Kingdom']
columns_to_select = ['date_dt', 'country_region', 'confirmed_delta']

data_countries = data_final.loc[  # создадим таблицу со странами из countries
    data_final['country_region'].isin(countries), columns_to_select  # rows, columns
].reset_index(drop=True)

data_countries.insert(0, 'year', data_countries['date_dt'].dt.year)

data_countries.head()
```

Переведем ее в wide формат, причем так, чтобы вместо колонки `country_region` стало три колонки, соответствующие каждой стране, со значениями из переменной `confirmed_delta`:

```python
data_countries.pivot(
    columns=['country_region'], index=['year', 'date_dt'], values=['confirmed_delta']
)
```

```python
def extract_index(index_tuple: tuple) -> str:
    first, second = index_tuple

    if not second:
        return first

    return second
```

```python
data_countries_pivoted = data_countries.pivot(
    columns=['country_region'], index=['year', 'date_dt'], values=['confirmed_delta']
).reset_index()

data_countries_pivoted.columns = data_countries_pivoted.columns.map(extract_index)

data_countries_pivoted.head()
```

Обратите внимание, что `pivot` работает только в том случае, если на каждую комбинацию значений индексных колонок приходится ровно одно значение из колонок значений (аргумент `values`):

```python
data_countries.pivot(
    columns=['country_region'], index=['year'], values=['confirmed_delta']
)
```

Что же делать в случае, когда на  такую комбинацию приходится более одного значения? Только агрегировать.

**Задача.** Получите сводную таблицу, в которой:

- колонки - год и страны;
- в ячейках средние значения переменной `confirmed_delta` по году и стране.

```python
data_countries.pivot_table(
    columns=['country_region'], index=['year'], values=['confirmed_delta']
)
```

А теперь вернемся обратно к long-формату. Для этого нам пригодится `melt`:

```python
data_countries_pivoted.melt(
    id_vars=['year', 'date_dt'], value_vars=['Russia', 'US', 'United Kingdom'],
    var_name='country_region', value_name='confirmed_delta'
)
```

## Разница между `groupby` и `pivot_table`
`pivot` в отличие от `pivot_table` не агрегирует, так что с ним все понятно.

`groupby` обычно используют, когда нужно разделить строки таблицы на группы по набору колонок, чтобы:

- применить агрегацию (с помощью агрегирующей функции) к другим переменных (колонкам) в каждой из этих групп;
- применить трансформацию группы (`pd.DataFrame` или `pd.Series`).

Что именно агрегируется и что получается на выходе? Это определяется типом аргументов:

```python
[*data_final.groupby('country_region')][:2]  # строка -> в кортежах первый элемент скалярный
```

```python
[*data_final.groupby(['country_region'])][:2]  # список -> в кортежах первый элемент - вектор
```

Обратите внимание, что группы (в данном случае страны) становятся идентификаторами строк.

`pivot_table` обычно применяют, когда главная цель - привести таблицу к другому формату. Агрегация здесь второстепенна и используется постольку, поскольку в ней есть необходимость.

```python
data_final.pivot_table(columns=['country_region'], values=['confirmed_delta'], aggfunc='mean')
```

Тут уже названия стран становятся названиями колонок, не строк.

## Как проводить join'ы в Pandas
Создадим демонстрационные данные:

```python
import numpy as np
import pandas as pd

from uuid import uuid1

# таблица клиентов
clients = pd.DataFrame(
    {
        'id': np.random.permutation(np.arange(10)),                    # идентификатор клиента
        'age': np.random.normal(40, 15, 10).round().astype(int),       # возраст клиента
        'gender': np.random.choice(['F', 'M'], size=10, replace=True)  # пол клиента
    }
)

# сессии клиентов (клиент заходит на некоторый сайт)
sessions = pd.DataFrame(
    {
        'session_id': [str(uuid1()) for i in range(30)],                             # идентификатор сессии
        'client_id': np.random.choice([1, 24, 5, 17, 8, 9], size=30, replace=True),  # идентификатор клиента
        'duration': np.random.normal(180, 50, 30).round().astype(int)                # продолжительность сессии (в секундах)
    }
)

display(
    clients.head(),
    sessions.head()
)
```

**Немного про виды связей** (подробнее см. в вспомогательных материалах)<br>
Их существует 3:

1. Один к одному (1:1) - каждая запись в одной таблице связана с одной и только одной записью в другой таблице. Пример: муж и жена (в стандартном случае);
2. Один к многим (1:n) - каждая запись в одной таблице связана с несколькими записями в другой таблице, но каждая запись в другой таблице связана только с одной записью в первой таблице. Пример: мать и ребенок;
3. Многие к многим (m:n) - каждая запись в одной таблице может быть связана с несколькими записями в другой таблице, и наоборот. Пример: родитель и ребенок.

```python
clients_set = set(clients['id'].unique())
sessions_set = set(sessions['client_id'].unique())
```

```python
clients_set
```

```python
sessions_set
```

**inner join** - пересечение ключей из левой и правой таблиц:

![](https://www.thoughtco.com/thmb/xh4MUu8HQyX1JVEcxn2IorWogoo=/1500x0/filters:no_upscale():max_bytes(150000):strip_icc()/0iHiL-d7a0c49a861448cb94477386a6f3f05b.png)

```python
intersection = clients_set & sessions_set
```

```python
display(
    sessions.head(),
    clients.head(),
)
```

```python
sessions_clients_inner_joined = sessions.merge(
    clients,
    left_on=['client_id'],
    right_on=['id'],
    how='inner',
    validate='m:1'  # удостоверимся, что связь действительно вида "многие к одному"
)
sessions_clients_inner_joined
```

```python
intersection == set(sessions_clients_inner_joined['client_id'])
```

**left join** - все ключи из левой таблицы и соответствующие им из правой:

```python
sessions_clients_left_joined = sessions.merge(
    clients,
    left_on=['client_id'],
    right_on=['id'],
    how='left'
)
sessions_clients_left_joined
```

```python
sessions_set == set(sessions_clients_left_joined['client_id'])
```

```python
sessions_set > set(sessions_clients_left_joined['id'].dropna().astype(int))
```

**right join** - все ключи из правой таблицы и соответствующие им из левой:

```python
sessions_clients_right_joined = sessions.merge(
    clients,
    left_on=['client_id'],
    right_on=['id'],
    how='right'
)
sessions_clients_right_joined
```

```python
clients_set == set(sessions_clients_right_joined['id'])
```

```python
clients_set > set(sessions_clients_right_joined['client_id'].dropna().astype(int))
```

**full join** - объединение ключей из левой и правой таблиц:

```python
union = clients_set | sessions_set
```

```python
sessions_clients_outer_joined = sessions.merge(
    clients,
    left_on=['client_id'],
    right_on=['id'],
    how='outer'
)
sessions_clients_outer_joined
```

```python
union == set(sessions_clients_outer_joined['client_id'].dropna().astype(int)) | set(sessions_clients_outer_joined['id'].dropna().astype(int))
```
